## 最小二乘法不带正则项与带正则项的拟合效果比较

```python
train_x, train_y, x, y = data.generate_data(9, 10)
lsm_fit = data.lsm(train_x, train_y)
lsm_punish = data.lsm(train_x, train_y, np.exp(-9))
plot1 = plt.plot(x, y, 'co', label='train_data')
real_x = np.linspace(0, 1, 50)
real_y = np.sin(real_x * 2 * np.pi)
lsm_fit_y = lsm_fit(real_x)
lsm_punish_y = lsm_punish(real_x)
plot2 = plt.plot(real_x, lsm_fit_y, 'b', label='lsm_fit')
plot3 = plt.plot(real_x, lsm_punish_y, 'r', label='lsm_punish_fit')
plot4 = plt.plot(real_x, real_y, 'g-.', label='real_data')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc=3)
plt.title("Compare with and without punishment, Order=9, Sample_size=10")
plt.show()
```
